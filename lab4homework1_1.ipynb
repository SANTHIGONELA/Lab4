```
import string
p=string.punctuation+string.whitespace
def load_word_as_dict(fn):
	d = {}
	with open(fn) as fin:
		for line in fin:
			if not line.strip().startswith('#'):
				for word in line.split():
					new_word = word.translate({ord(c): None for c in (p)}).lower()
					if new_word not in d:
						d[new_word] = 1
					else:
						d[new_word] += 1
	return d
def check_word_in_book(wordlist, book_wordlist):
	for word in book_wordlist.keys():
		if word not in wordlist.keys():
			print(word)
d1 = load_word_as_dict('word.txt')
d2 = load_word_as_dict('mybook.txt')
check_word_in_book(d1, d2)
```

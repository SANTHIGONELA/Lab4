```
import string
p=string.whitespace+string.punctuation
fin=open("mybook.txt")
#data=fin.read()
wf=[]
for line in fin:
  line=line.split()
  
  for word in line:
     word=word.strip(p)
     word=word.lower()
     wf.append(line.count(word))
     #print(word)
     #print(len(word))
#print("the total number of books":len(data))
     print("Pairs\n" + str(list(zip(line, wf))))
```

```
import string
def r_file(fn):
	remove_chars = list(string.punctuation + string.whitespace)
	with open(fn) as fin:
		for line in fin:
			if not line.strip().startswith('#'):
				for word in line.split():
					yield word.translate({ord(x): '' for x in remove_chars}).lower()

def subtraction(s1, s2):
	return s2 - s1
					

s2 = set(r_file('word.txt'))
s1 = set(r_file('mybook.txt'))
print (subtraction(s1, s2))
print(len(subtraction(s2,s1)))
```
